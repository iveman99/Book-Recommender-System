## Model on Popularity based recomendation system

```python
Rating_With_Bookname = Ratings.merge(Books,on='ISBN')
```

```python
Rating_With_Bookname 
```

```python
Numberof_Rating_df = Rating_With_Bookname.groupby('Book-Title').count()['Book-Rating'].reset_index()
Numberof_Rating_df.rename(columns={'Book-Rating':'Number_of_Ratings'},inplace=True)
Numberof_Rating_df
```

```python
# Convert Book-Rating to numeric (coerce errors to NaN)
Rating_With_Bookname['Book-Rating'] = pd.to_numeric(Rating_With_Bookname['Book-Rating'], errors='coerce')

# Now group and calculate mean
AVG_Rating_df = Rating_With_Bookname.groupby('Book-Title')['Book-Rating'].mean().reset_index()
AVG_Rating_df.rename(columns={'Book-Rating': 'Average_Ratings'}, inplace=True)
AVG_Rating_df
```

```python
Popular_DF = Numberof_Rating_df.merge(AVG_Rating_df,on='Book-Title')
Popular_DF
```

```python
Popular_DF[Popular_DF['Number_of_Ratings'] >= 250]
```

```python
Popular_DF[Popular_DF['Number_of_Ratings'] >= 250].sort_values('Average_Ratings',ascending = False)
```

```python
Popular_DF = Popular_DF[Popular_DF['Number_of_Ratings'] >= 250].sort_values('Average_Ratings',ascending = False).head(50)
```

```python
Popular_DF
```

```python
Popular_DF.merge(Books,on = 'Book-Title').drop_duplicates('Book-Title').shape
```

## Model on Collebrating Filtering Based Recomendation System

```python
x= Rating_With_Bookname.groupby('User-ID').count()['Book-Rating'] >200
```

```python
Rated_Users = x[x].index
Rated_Users
```

```python
Filtered_Rating = Rating_With_Bookname[Rating_With_Bookname['User-ID'].isin(Rated_Users)]
```

```python
y = Filtered_Rating.groupby('Book-Title').count()['Book-Rating'] >=50
Famous_Books= y[y].index
```

```python
Famous_Books.shape
```

```python
Filtered_Rating = Filtered_Rating[Filtered_Rating['Book-Title'].isin(Famous_Books)]
```

```python
pt = Filtered_Rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_Score=cosine_similarity(pt)
```

```python
Similarity_Score.shape
```

```python
def recommend(Book_name):
    # index fetch
    index = np.where(pt.index == Book_name)[0][0]
    
    # sort the books based on similarity score
    similar_items = sorted(list(enumerate(Similarity_Score[index])), key=lambda x: x[1], reverse=True)[1:6]
    
    # print the indices of top 5 similar books
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('The Notebook')
```
